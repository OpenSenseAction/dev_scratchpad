```python
import numpy as np
import xarray as xr
```

```python
ds_points = xr.Dataset(
    data_vars={'rainfall_amount': (('id', 'time'), np.random.rand(4, 20))},
    coords={
        'id': ['p1', 'p2', 'p1_same_loc', 'p1_same_x'],
        'time': np.arange(20),
        'x': ('id', [1, 2, 1, 1]),
        'y': ('id', [4, 3, 4, 9]),
    },
)
ds_points
```

```python
df = ds_points.id.to_dataframe()
duplicated_xy = df.duplicated(['x', 'y'], keep=False)
duplicated_xy
```

```python
da_duplicated_xy = xr.DataArray(duplicated_xy)
da_duplicated_xy
```

```python
ds_points.isel(id=~da_duplicated_xy)
```
